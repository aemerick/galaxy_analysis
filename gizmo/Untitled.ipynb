```python
plot_settings()

def plot_scatters(data_dict, outname, show_extra = None):
    
    fig,all_ax = plt.subplots(2,2)
    fig.set_size_inches(12,12)
    elements = ['C','N','O','Ne','Mg','Fe','Si','S','Ca']
    axi = 0
    num = np.arange(np.size(elements))

    ax = all_ax[0]
    for denom in ['Fe','O']:

        sig2 = np.zeros(np.size(elements))
        sig3 = np.zeros(np.size(elements))


        for i,e in enumerate(elements):

            if e == denom:
                sig2[i] = -1.0
                sig3[i] = -1.0
            else:
                field = '[' + e + '/' + denom + ']'
                sig2[i] = data_dict[field]['star']['2-sigma']
                sig3[i] = data_dict[field]['star']['3-sigma']            

        ax[axi].scatter(num, sig2, color = 'C0', s = 50, marker = 'o')
        ax[axi].scatter(num, sig3, color = 'C1', s = 50, marker = 'o')

        ax[axi].plot(num,sig2,color = 'C0', lw = 2, ls = '-', label = '2-sigma')
        ax[axi].plot(num,sig3,color = 'C1', lw = 2, ls = '-', label = '3-sigma')

        ax[axi].set_ylabel("[X/"+denom+"] Scatter [dex]")

        axi = axi + 1

        
    if show_extra:
        axi = 0
        for denom in ['Fe','O']:

            sig2 = np.zeros(np.size(elements))
            sig3 = np.zeros(np.size(elements))
            for i,e in enumerate(elements):

                if e == denom:
                    sig2[i] = -1.0
                    sig3[i] = -1.0
                else:
                    field = '[' + e + '/' + denom + ']'
                    sig2[i] = show_extra[field]['star']['2-sigma']
                    sig3[i] = show_extra[field]['star']['3-sigma']            

            ax[axi].scatter(num, sig2, color = 'C0', s = 50, marker = 'D')
            ax[axi].scatter(num, sig3, color = 'C1', s = 50, marker = 'D')

            ax[axi].plot(num,sig2,color = 'C0', lw = 2, ls = '--')#, label = '2-sigma')
            ax[axi].plot(num,sig3,color = 'C1', lw = 2, ls = '--')#, label = '3-sigma')
            
    #-----------------
            axi = axi + 1
    

    ax = all_ax[1]
    axi = 0
    for denom in ['Fe','O']:

        sig2 = np.zeros(np.size(elements))
        sig3 = np.zeros(np.size(elements))


        for i,e in enumerate(elements):

            if e == denom:
                sig2[i] = -1.0
                sig3[i] = -1.0
            else:
                field = '[' + e + '/' + denom + ']'
                sig2[i] = m12s[field]['gas']['2-sigma']
                sig3[i] = m12s[field]['gas']['3-sigma']            

        ax[axi].scatter(num, sig2, color = 'C0', s = 50, marker = 'o')
        ax[axi].scatter(num, sig3, color = 'C1', s = 50, marker = 'o')

        ax[axi].plot(num,sig2,color = 'C0', lw = 2, ls = '-', label = '2-sigma')
        ax[axi].plot(num,sig3,color = 'C1', lw = 2, ls = '-', label = '3-sigma')

        ax[axi].set_ylabel("[X/"+denom+"] Scatter [dex]")


        axi = axi + 1
        
        
    if show_extra:
        axi = 0
        for denom in ['Fe','O']:

            sig2 = np.zeros(np.size(elements))
            sig3 = np.zeros(np.size(elements))
            for i,e in enumerate(elements):

                if e == denom:
                    sig2[i] = -1.0
                    sig3[i] = -1.0
                else:
                    field = '[' + e + '/' + denom + ']'
                    sig2[i] = show_extra[field]['star']['2-sigma']
                    sig3[i] = show_extra[field]['star']['3-sigma']            

            ax[axi].scatter(num, sig2, color = 'C0', s = 50, marker = 'D')
            ax[axi].scatter(num, sig3, color = 'C1', s = 50, marker = 'D')

            ax[axi].plot(num,sig2,color = 'C0', lw = 2, ls = '--')#, label = '2-sigma')
            ax[axi].plot(num,sig3,color = 'C1', lw = 2, ls = '--')#, label = '3-sigma')
            
    #-----------------
            axi = axi + 1        
        
    all_ax[(0,0)].legend(loc='upper right')

    all_ax[(0,0)].annotate("Stars",(0.05,0.92), (0.05,0.92),xycoords='axes fraction')
    all_ax[(1,0)].annotate("Gas",(0.05,0.92), (0.05,0.92),xycoords='axes fraction')

    for ax in all_ax:
        for axi in [0,1]:
            ax[axi].set_yticks(np.linspace(0,1.0,11))
            ax[axi].set_xticks(num)
            ax[axi].set_xticklabels(elements)
            ax[axi].set_ylim(0.0,1.0)
            #ax[axi].p

    plt.tight_layout()
    fig.savefig(outname)

    return

plot_scatters(m12s, "m12_average_scatter.png", show_extra = nugrid)
                    #show_extra_label = 'NuGrid', title = "M12")
plot_scatters(m10s, "m10_average_scatter.png", title = "m10s")

plot_scatters(nugrid, "nugrid_average_scatter.png", title = "Nugrid-M12"
```
